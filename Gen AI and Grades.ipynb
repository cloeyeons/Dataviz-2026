```python
areas = ['Health', 'STEM', 'Social Sciences', 'Arts and Humanities']

sa = np.array([21, 59, 27, 4])
agr = np.array([53, 116, 79, 30])
nei = np.array([41, 97, 85, 28])
dis = np.array([29, 82, 88, 34])
sd = np.array([13, 22, 40, 16])
na = np.array([4, 14, 8, 5])

data = pd.DataFrame({
    'Strongly Agree': sa,
    'Agree': agr,
    'Neither Agree nor Disagree': nei,
    'Disagree': dis,
    'Strongly Disagree': sd,
    "I Don't Know/NA": na
}, index = areas)

data_prop = data.div(data.sum(axis = 1), axis = 0)

data = data_prop.sort_values(by = 'Strongly Agree')

# Plot horizontal stacked bar chart
data.plot(kind = 'barh', stacked = True, figsize = (8,5), color = colors)

# Labels and formatting
plt.xlabel('Proportion')
plt.ylabel('Area of Study')
plt.title('Anticipation by Area of Study', pad = 20)
plt.legend(loc = 'lower right')
plt.show()
```
